```python
from IPython.display import Markdown as md, display

notes = '''**from boolean to conditional expressions**
--
first, booleans are *not conditions* themselves but they
are *used in conditions*

**what are booleans?**
- a boolean is a *data type* in python that can have
only one of two values
    - `true`
    - `false`
- they represent the result of a **logical test** or a 
condition
- python uses these to decide whether a block of code should run 
or not

**how booleans and conditions work together**
- when you write a conditional statement like `if`,
python checks the condition (logical statement)
and gets a boolean (`true` or `false`)

'''
display(md(notes))

# booleans and conditions working together
x = 10
if x > 5: 
    print("x is greater than 5")
else:
    print("x is not greater than 5")

# booleans by themselves
is_raining = True

if is_raining:
    print("Bring an unbrella!")
else:
    print("No umbrella needed!")


notes2 = '''**comparison between boolean and conditionals**
--
**booleans**
- a data type (`True` or `False`)
- booleans *result from* conditionals

**conditionals**
- a *logical statement* that produces a boolean
- conditionals *produce* booleans 

'''
display(md(notes2))

notes3 = '''
let's think of **conditional statements** like 
forks in the road where you tell the computer what to do
depending on which path it takes

what are **conditional statements?
- are like saying
    - if this happens, do this
    - otherwise do something else

*example:*
'''

display(md(notes3))

def check_number(number): 
    if number > 0:
        return "This is a positive number!"
    elif number < 0:
        return "This is a negative number!"
    else:
        return "This is zero!"

print(check_number(0))
```

```python
from IPython.display import Markdown as md, display
notes = '''****chained conditionals****
--
**chained conditionals** let your program choose bw 
*more than two options.* it's like a chain of *if this, do that.
otherwise, check something else*

**real life example**
- imagine you're deciding what to wear based on temp
    - if its below 32 degrees, wear a jacket
    - if it's above 32 and below 60 degrees, wear a sweater
    - if it's above 60 degrees, wear a t-shirt


'''

display(md(notes))

def what_to_wear(temperature):
    if temperature < 32:
        return "Wear a jacket!"
    elif temperature <= 60:
        return "Wear a sweater!"
    else:
        return "Wear a t-shirt!"

print(what_to_wear(50))

notes2 = '''**how it works**
---
- the first `if` checks if the temp is below 32 so 
if true, it stops there
- if the first condition is false, the `elif` checks if the 
temperature is bw 32 and 60 so if true, it stops there
- if neither conditions are true, the `else` runs, meaning the
temp is above 60

*more complex example: grading system*
'''

display(md(notes2))

def grade(score):
    if score >= 90:
        return "A"
    elif score >=80:
        return "B" 
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else: 
        return "F"

print(grade(95))

notes3 = '''**the difference bw conditionals and chained**
--
**regular conditionals**
- involve a *single conditional* or a simple *if-else*
- like deciding between just two paths

**chained conditionals**
- involve *multiple conditions* that 
check *one after the other*
- deciding on many paths 
    - if one condition fials, move on to the next one
'''

display(md(notes3))
```

```python
from IPython.display import Markdown as md, display

notes ='''**conditions can be nested**
--
a **nested conditional** is when you put one conditional
statement (like `if`, `elif`, or `else`) *inside another
conditional statement*. it's like having a decision 
inside another decision

**how it works**
--
- the *outer conditional* checks the first condition
- if the outer condition is true, the program goes
*into another conditional* inside the first one to check further


**real-life examples:**
--
imagine you're deciding whether to go outside:
- if it's sunny, you'' decide to wear:
    - if it's warm, you'll wear a t-shirt
    - if it's cold, you'll wear a jacket
- if it's not sunny, you'll stay home

'''

display(md(notes))

def what_to_do(weather, temperature):
    if weather == "sunny": # outer conditional
        if temperature > 60: # inner conditional
            return "Wear a t-shirt!"
        else:
            return "Wear a jacket!"
    else:
        return "Stay home!"

print(what_to_do("rainy", 50))

notes2 = '''**explanation**
--
- **the outer conditional** checks if the temp is `sunny`
    - if true, the program checks the *inner conditional* 
- if the **outer condition** is false (not sunny), the 
program skips the inner conditions and goes straight to
`stay home`
    
*example with grading*
'''

display(md(notes2))

def grade_with_honors(score):
    if score >= 60: # outer conditional
        if score >= 90: # inner conditional
            return "You passed with honors!"
        else:
            return "You passed!"
    else:
        return "You failed."

print(grade_with_honors(90))
```

```python
from IPython.display import Markdown as md, display

notes = '''**what is string indexing?**
--
- **string indexing** lets you access individual characters
in a string using their position (index)
- each character in a string has a specific **index number** starting
from `0` for their first character

**position indexing**
--
s = "hello"
if you say: `print(s[0]) it will ouput 'h' 
since it's the first letter

**negative indexing**
--
- negative indexing starts from -1 for the last character
and moves backwards

example:
*s = "hello"*

print(s[-1]) would output o for the last letter

print(s[-2]) would output l for the second the last character

**accessing substring**
--
- you can use *slicing* to extract multiple characters
(substring)
- the format is `string[start:stop:step]`
    - `start` is the index to start at (inclusive)
    - `stop` is the index to stop at (exclusive)
    - `step` is the step size (defaults at 1)

*example*
s = "hello"

print(s[1:4]) = "ell"

print(s[:3]) = "hel"

print(s[2:] = "llo"

print(s[::2] = "hlo" 

print(s[::-1] = "olleh"

**key points**
--
- indexing is 0-based so first character has index of `0`
- use negative indices to count from the end
- *slicing* allows you to extract substrings
- trying to access outside the range = **IndexError**
'''

display(md(notes))
```
